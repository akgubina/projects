```python
#from funding rounds select companies that had B-rounds
#some companies have multiple B-rounds, let's keep only the earliest one
b_rounds = funding_rounds[funding_rounds.investment_type=='series_b'].sort_values(by=['org_name','announced_on'])
b_rounds = b_rounds[~b_rounds.duplicated('org_name',keep='first')]
```

```python
#retaing relevant columns in b_rounds
b_rounds = b_rounds[['uuid','org_uuid','lead_investor_uuids','raised_amount_usd', 'announced_on']]
```

```python
#join v1 and b_rounds to get v2
v2 = pd.merge(organizations,b_rounds, how='left', on='uuid')
```
